```python
n=int(input("Enter a number:"))
x=1
while(n>0 and x<=10):
    mul=n*x
    print(n,"*",x,"=",mul)
    x+=1
```

```python
lines = 4
for i in range(lines):
    spaces = lines - i - 1
    stars = 2 * i + 1
    print(" " * spaces + "*" * stars)
```
