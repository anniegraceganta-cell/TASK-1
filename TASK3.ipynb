```python
import pandas as pd
import numpy as np

data = {
    'Student_ID': [101, 102 , 103, 104, 105, 105,106],
    'Name': ['Alice ', 'bob', 'Charlie', 'Daisy', 'Ethan', 'Ethan', 'Frank'], # Typos/spaces
    'Subject': ['Math', 'Science', 'Math', 'History', 'Science', 'Science', 'UNKNOWN'], # Missing value string
    'Score': [85.0, np.nan, 92.0, -10.0, 45.0, 45.0, 88.0], # NaN and an impossible negative score
    'Passed': ['True', 'False', 'TRUE', 'True', 'False', 'False', 'UNKNOWN'] # Inconsistent text
}

df_data = pd.DataFrame(data)
df_data
```

```python
#removing duplicates
df_data = df_data.drop_duplicates()
```

```python
#cleaning string
df_data['Name'] = df_data['Name'].str.strip().str.capitalize()
```

```python
#fixing outliers
df_data.loc[df_data['Score'] < 0, 'Score'] = np.nan
```

```python
#handling missing values
mean_score = df_data['Score'].mean()
df_data['Score'] = df_data['Score'].fillna(mean_score)
```

```python
#standardizing text values
df_data['Passed'] = df_data['Passed'].astype(str).str.upper()
df_data['Passed'] = df_data['Passed'].replace('UNKNOWN', 'FALSE')
```

```python
#replace specific text errors
df_data['Subject'] = df_data['Subject'].astype(str).str.upper()
df_data['Subject'] = df_data['Subject'].replace('UNKNOWN', 'ENGLISH')
```

```python
median_value = df_data['Student_ID'].median()
#FillING the missing cells with that median
df_data['Student_ID'] = df_data['Student_ID'].fillna(median_value)
df_data
```

```python
median_value = df_data['Student_ID'].median()
# 2. Fill the missing cells with that median
df_data['Student_ID'] = df_data['Student_ID'].fillna(median_value)
df_data
```
